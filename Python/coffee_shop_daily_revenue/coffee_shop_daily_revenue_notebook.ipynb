```python
# Import libraries

## Data manipulation
import pandas as pd

## Data visualisation
import matplotlib.pyplot as plt
import seaborn as sns

## Machine learning
```

```python
# Import the dataset
coffee = pd.read_csv("coffee_shop_daily_revenue_dataset.csv")
```

```python
# Preview the data
coffee.head(10)
```

```python
# View the info
coffee.info()
```

Comment: There is no missing values.

```python
# Get summary statistics
coffee.describe()
```

```python
# Visualise the distribution of each variable

## Get all column names
col_list = coffee.columns

## Set up a for loop
for col in col_list:
    
    ## Create a histogram
    sns.histplot(x=col,
                 data=coffee,
                 kde=True)
    
    ## Add a title and labels
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    
    ## Show the plot
    plt.show()
```

Comments:
- All variables, except `Daily_Revenue`, have a uniform distribution.
- `Daily_Revenue` appears normally distribution despite being slightly positively skewed.

```python
# Visualise outliers

## Set up a for loop
for col in col_list:
    
    ## Create a histogram
    sns.boxplot(x=col,
                data=coffee)
    
    ## Add a title and labels
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    
    ## Show the plot
    plt.show()
```

```python
# Create a correlation matrix

## Get a correlation matrix
cor_mat = coffee.corr()

## Plot the correlation matrix
sns.heatmap(data=cor_mat,
            annot=True,
            cmap="coolwarm",
            fmt=".2f",
            linewidths=.5)

## Add a title
plt.title("Correlation Matrix")

## Customise x labels
plt.xticks(rotation=45,
           ha="right")
plt.tight_layout()

## Show the plot
plt.show()
```

Comments:
- Two strongest correlates of daily revenue are (1) the number of customers (r = 0.74) and (2) average order value (r = 0.54).
- This suggests that these two factors are major contributors to daily revenue.
- Another interest factor is the amount of money spent on marketing with correlation of 0.25.
- The other factors show extremely low to no correlation with daily revenue.
